## Mapping "Region" Values with numeric data.

```python
unique_regions = df['Region'].unique()
region_mapping = {region: idx for idx, region in enumerate(unique_regions)}
print("Region Mapping:", region_mapping)
df['Region'] = df['Region'].map(region_mapping)
df.head()
```

## from below graph it can be analyzed that ragion 10 has outlying value which is highest and it corresponds to North America as per mapping with array.

```python
plt.figure(figsize=(15, 10))
sns.scatterplot(x=df['Region'], y=df['combined figures (kg/capita/year)'])
```

### Based on the analysis, it is evident that the 'Combined Figures (kg/capita/year)', which represents the total amount of food waste produced per capita annually, encompassing waste from households, retail stores, and food service establishments, is highest in the <span style="background-color: yellow;">North America region on per capita basis</span>.
